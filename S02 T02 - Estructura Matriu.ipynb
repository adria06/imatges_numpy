## Nivell 3
**Manipulació d’imatges amb Matplotlib.**

I select an image where testing the removal of one color can easily be tested 

```python
%matplotlib inline
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
img = mpimg.imread('RGBColors.jfif')
imgplot = plt.imshow(img)
plt.axis('off')
```

Now we remove see the effect of removing one of the colors at a time.

```python
def subplot (img, name):             # Function to subplot 3 images each withour one of the colors RGB
    for i in range(3):
        plt.figure(figsize=(15,15))
        plt.subplot(1,3,i+1)
        plt.axis('off')
        img_copy=img.copy()
        img_copy[:, :,[i]]=0
        plt.imshow(img_copy)
        if i==0: plt.title('Image without R')
        elif i==1: plt.title('Image without G')
        else: plt.title('Image without B')
        file_name = name + str(i)+ '.png'
        plt.imsave(file_name, img_copy)

name = 'Triangulo'
subplot(img, name)
```

Now that the removal of the colors have been tested we can try the effect in other images.

Image 1: *Blue Buttefly*

```python
# after print(img2) we noted that it has a range 0 to 255, so it's necessary divided by 255 to have a range 0 to 1
img2 = mpimg.imread('butterfly.jfif')/255      
imgplot = plt.imshow(img2)
plt.axis('off')
```

*Blue butterfly removing RGB colors*

```python
name = 'Butterfly'
subplot(img2, name)
```
